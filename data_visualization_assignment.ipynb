```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
```

```python
import pandas as pd
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])
```

```python
features = ['sepal length','sepal width','petal length','petal width']
x=df.loc[:,features].values
y=df.loc[:,['target']].values
df.head()
```

```python
pca = PCA(n_components=3)
pc = pca.fit_transform(x)
pdf = pd.DataFrame(data=pc,columns=['pc1','pc2','pc3'])
pdf
```

```python
fin=pd.concat([pdf,df[['target']]],axis=1)
fin
```

```python
# using plotly for plotting 3d graphs is easy and give more info

import plotly.express as px
fig = px.scatter_3d(fin, x='pc1', y='pc2', z='pc3',
              color='target',title='3D SCATTER PLOT OF IRIS DATA',size_max=18)
fig.show()
```
