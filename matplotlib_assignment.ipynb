```python
import matplotlib.pyplot as plt

Max = [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
Min = [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]
x=[i for i in range(len(Max))]
plt.scatter(x,Max)
plt.scatter(x,Min)
```

```python
from scipy import optimize
def year_tem(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(year_tem, x,
                                      Max, [40, 20, 0])
res_min, cov_min = optimize.curve_fit(year_tem, x,
                                      Min, [-40, 20, 0])


days = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(x, Max, 'go')
plt.plot(days, year_tem(days, *res_max), 'm-')
plt.plot(x, Min, 'co')
plt.plot(days, year_tem(days, *res_min), 'y-')
plt.xlabel('Month')
plt.ylabel('Temperature in celsius')

plt.show()
```

```python
import pandas as pd 
import numpy as np
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
```

```python
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
x = titanic['sex'].value_counts(normalize=True)*100
```

```python
fig = plt.figure(figsize =(10, 7)) 
plt.pie(x, labels = ['male','female'],autopct='%1.2f%%')  
# show plot 
plt.show() 
```

```python
male_data = titanic[titanic['sex']=='male']
female_data = titanic[titanic['sex']=='female']

fig = plt.figure(figsize =(10, 7))
plt.scatter(male_data['age'],male_data['fare'],color='red',label='male')
plt.scatter(female_data['age'],female_data['fare'],color='green',label='female')
plt.legend(['male','female'])
plt.xlabel('AGE')
plt.ylabel('FARE')
```
